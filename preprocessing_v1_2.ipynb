### Read_

```python
df = pd.read_csv("/Users/2020/Housing-Market_Price_Project/preprocessing_apt_price.csv")
df_comm = pd.read_csv("/Users/2020/Housing-Market_Price_Project/seoul_commercial.csv", encoding = 'cp949')
df_csi = pd.read_excel("/Users/2020/Housing-Market_Price_Project/CSI.xls", header=1)
df_stat = pd.read_excel("/Users/2020/Housing-Market_Price_Project/stat_108001.xls", header=1)
df_call_ir = pd.read_csv("/Users/2020/Housing-Market_Price_Project/ECOS_TABLE_20211011_144110_시장금리(콜).csv", encoding = 'cp949')
```

### Key 만들기
연+분기 : 'year_quarter' : YYYYQq (type : str)<br>
연+월 : 'year_month' : YYYY-MM-01 (type : datetime64)

```python
df
```

```python
# df : 2010Q1 ~ 2017Q4 (type : str)
df['transaction_year_month'] = pd.to_datetime(df['transaction_year_month'], format = "%Y%m")
df.insert(0, 'year_quarter', df['transaction_year_month'].dt.year.astype(str) + \
          'Q' + df['transaction_year_month'].dt.quarter.astype(str))

# df
```

```python
df
```

```python
# df_comm : 2014Q1 ~ 2021Q2 (type : str)
temp_list = [0, 2000, 4000, 5000, 6000, 7500, 9000, 11000,\
             12500, 14000, 16000, 17000, 18000, 19500, 21000,\
             23000, 24000, 26000, 27000, 29000, 30000, 32000,\
             33000, 35000, 37000, 38000, 39000, 41000, 43000, 44000]
df_comm = df_comm.iloc[temp_list][::-1]
df_comm = df_comm[['기준_년_코드', '기준_분기_코드',\
                  '서울_운영_영업_개월_평균', '서울_폐업_영업_개월_평균']]
df_comm.rename(columns={'기준_년_코드':'year', '기준_분기_코드':'quarter'}, inplace=True)
df_comm['year'] = df_comm['year'].astype(str)
df_comm['quarter'] = df_comm['quarter'].astype(str)
df_comm.insert(0, 'year_quarter', df_comm['year'] + 'Q' + df_comm['quarter'])
df_comm.drop(['year', 'quarter'], axis=1, inplace=True)

# df_comm
```

```python
# df_csi : 2009Q1 ~ 2019Q4 (type : str)
df_csi.insert(0, 'year_quarter', df_csi['기간'].str[:4] + 'Q' + df_csi['기간'].str[-3])
df_csi.drop(['기간', '구분 '], axis=1, inplace=True)

# df_csi
```

```python
# df_stat : 2008-01-01 ~ 2021-08-01, freq : M, type:datetime64
df_stat = df_stat.transpose().iloc[85:,:5]
columns_df_stat = {0:'transaction_year_month',
                   1:'코스피지수',
                   2:'코스닥지수',
                   3:'코스피200',
                   4:'코스피50'
                  }
df_stat.rename(columns=columns_df_stat, inplace=True)
df_stat['transaction_year_month'] = df_stat['transaction_year_month'].str[:-1]
df_stat['transaction_year_month'] = pd.to_datetime(df_stat['transaction_year_month'], format = "%Y%m")
df_stat['코스피지수'][:-2] = df_stat['코스피지수'][:-2].str.replace(',', '')
df_stat['코스닥지수'][:-2] = df_stat['코스닥지수'][:-2].str.replace(',', '')
df_stat
```

```python
df_call_ir = df_call_ir.transpose().reset_index().iloc[4:, :2]
df_call_ir['index'] = df_call_ir['index'].str.replace('/', '')
```

```python
df_call_ir['index'] = df_call_ir['index'].str.replace('/', '')
df_call_ir['index'] = pd.to_datetime(df_call_ir['index'], format = "%Y%m")
```

```python
df_call_ir.rename(columns={'index':'transaction_year_month', 0:'시장금리(무담보콜)'}, inplace=True)
```

```python
df_call_ir
```

### Merge(left outer join)
left.merge(right, on='key', how='left')

```python
df = df.merge(df_comm, on='year_quarter', how='left')
df = df.merge(df_csi, on='year_quarter', how='left')
df = df.merge(df_stat, on='transaction_year_month', how='left')
df = df.merge(df_call_ir, on='transaction_year_month', how='left')

# pd.set_option('display.max_column', 1000000)
df
```
